# Construindo o modelo de Regressão Linear

## CONSIDERANDO O TEMPO COMO VARIÁVEL INDEPEDENTE
Nesse caso, consideramos que o tempo é a variável independente e cada eixo varia conforme o passar do tempo. Isso permite um modelo mais reativo com a espera do goleiro para tomar uma reação frente ao chute

## Equações:

Devemos estabelecer um modelo de regressão para cada eixo, baseado no tempo.
<br><br>
$X=$ $\alpha_1*t$ + $\beta_1$ + $\epsilon$<br>
$Y=$ $\alpha_2*t$ + $\beta_2$ + $\epsilon$<br>
$Z=$ $\alpha_3*t$ + $\beta_3$ + $\epsilon$

### Definindo as funções de custo, de pesos e de fit_linear

```python
def cost_function(T,X,alpha,beta):
    '''MSE   RMSE'''
    x_hat_list=[]
    total_error=0.0
    N=len(X)
    for i in range(N):
        x_hat = alpha * T[i] + beta
        x_hat_list.append(x_hat)
        total_error += (X[i] - x_hat )**2
    mse = total_error/N
    rmse = np.sqrt(mse)
    return mse , rmse , x_hat_list


def update_weights(T,X,alpha,beta,learning_rate):
    alpha_deriv=0
    beta_deriv=0
    N=len(X)
    for i in range(N):
        alpha_deriv += -2*T[i] * (X[i] - (alpha*T[i] + beta))
        beta_deriv += -2 * (X[i] - (alpha*T[i] + beta))
    alpha -= (alpha_deriv / float(N)) * learning_rate
    beta -= (beta_deriv / float(N)) * learning_rate
    return alpha, beta



def fit_linear(T, X, alpha, beta, learning_rate, iters):
    cost_history = []
    space_inter=iters/5
    mse_fim=0
    rmse_fim=0
    for i in range(int(iters)):
        alpha,beta = update_weights(T, X, alpha, beta, learning_rate)

        #Calculate cost for auditing purposes
        mse_ini,rmse_ini,x_hat = cost_function(T, X, alpha, beta)
        cost_history.append(rmse_ini)
        
        #Checking for the tolerance of the error
        tol_rmse = np.abs(rmse_ini-rmse_fim)
        tol_mse = np.abs(mse_ini-mse_fim)
        if tol_rmse < 1e-10:
            break
        mse_fim=mse_ini
        rmse_fim=rmse_ini
        
        # Log Progress
        if i % space_inter == 0:
            print ("iter={:d}    alpha={:.2f}    beta={:.4f}    MSE={:.2}     RMSE={:.2}".format(i, alpha, beta, mse_fim, rmse_fim))
            plt.scatter(T,X)
            plt.plot(T,x_hat,c='red')
            plt.show()
    return rmse_fim, alpha, beta, cost_history, i
```

## Equações:

Devemos estabelecer um modelo de regressão para cada eixo, baseado no tempo.
<br><br>
$X=$ $\alpha_1*t$ + $\beta_1$ + $\epsilon$<br>
$Y=$ $\alpha_2*t$ + $\beta_2$ + $\epsilon$<br>
$Z=$ $\gamma_3*t²$ + $\alpha_3*t$ + $\beta_3$ + $\epsilon$, where $\gamma_3$ = -9.8 m/s²

### Definindo as funções de pesos e de fit_polinomial_grav

```python
def cost_function_pol_grav(T,X,gamma,alpha,beta):
    '''MSE   RMSE'''
    x_hat_list=[]
    total_error=0.0
    N=len(X)
    for i in range(N):
        x_hat = gamma * T[i]**2 + alpha * T[i] + beta
        x_hat_list.append(x_hat)
        total_error += (X[i] - x_hat )**2
    mse = total_error/N
    rmse = np.sqrt(mse)
    return mse , rmse , x_hat_list


def update_weights_pol_grav(T,X,gamma, alpha,beta,learning_rate):
    alpha_deriv=0
    beta_deriv=0
    N=len(X)
    for i in range(N):
        alpha_deriv += -2*T[i] * (X[i] - (gamma * T[i]**2 + alpha*T[i] + beta))
        beta_deriv += -2 * (X[i] - (gamma * T[i]**2 + alpha*T[i] + beta))
    alpha -= (alpha_deriv / float(N)) * learning_rate
    beta -= (beta_deriv / float(N)) * learning_rate
    return alpha, beta



def fit_polinomial_grav(T, X, gamma, alpha, beta, learning_rate, iters):
    cost_history = []
    space_inter=iters/5
    mse_fim=0
    rmse_fim=0
    for i in range(int(iters)):
        alpha,beta = update_weights_pol_grav(T, X, gamma, alpha, beta, learning_rate)

        #Calculate cost for auditing purposes
        mse_ini,rmse_ini,x_hat = cost_function_pol_grav(T, X, gamma, alpha, beta)
        cost_history.append(rmse_ini)
        
        #Checking for the tolerance of the error
        tol_rmse = np.abs(rmse_ini-rmse_fim)
        tol_mse = np.abs(mse_ini-mse_fim)
        if tol_rmse < 1e-10:
            break
        mse_fim=mse_ini
        rmse_fim=rmse_ini
        
        # Log Progress
        if i % space_inter == 0:
            print ("iter={:d}  gamma={:.2f}  alpha={:.2f}    beta={:.4f}    MSE={:.2}     RMSE={:.2}".format(i,gamma, alpha, beta, mse_fim, rmse_fim))
            plt.scatter(T,X)
            plt.plot(T,x_hat,c='red')
            plt.show()
    return alpha, beta, cost_history, i
```

## 2) DEIXANDO O SISTEMA DEFINIR O GAMMA PARA FUNÇÃO QUADRÁDICA

## Equações:

Devemos estabelecer um modelo de regressão para cada eixo, baseado no tempo.
<br><br>
$X=$ $\gamma_1*t²$ + $\alpha_1*t$ + $\beta_1$ + $\epsilon$<br>
$Y=$ $\gamma_2*t²$ + $\alpha_2*t$ + $\beta_2$ + $\epsilon$<br>
$Z=$ $\gamma_3*t²$ + $\alpha_3*t$ + $\beta_3$ + $\epsilon$

### Definindo as funções de pesos e de fit_polinomial_quadr

```python
def cost_function_pol_quadr(T,X,gamma,alpha,beta):
    '''MSE   RMSE'''
    x_hat_list=[]
    total_error=0.0
    N=len(X)
    for i in range(N):
        x_hat = gamma * T[i]**2 + alpha * T[i] + beta
        x_hat_list.append(x_hat)
        total_error += (X[i] - x_hat )**2
    mse = total_error/N
    rmse = np.sqrt(mse)
    return mse , rmse , x_hat_list


def update_weights_pol_quadr(T,X,gamma, alpha,beta,learning_rate):
    
    gamma_deriv=0
    alpha_deriv=0
    beta_deriv=0
    
    N=len(X)
    for i in range(N):
        gamma_deriv += -2 *(T[i]**2) * (X[i] - (gamma * T[i]**2 + alpha*T[i] + beta))
        alpha_deriv += -2*T[i] * (X[i] - (gamma * T[i]**2 + alpha*T[i] + beta))
        beta_deriv += -2 * (X[i] - (gamma * T[i]**2 + alpha*T[i] + beta))
    gamma -= (gamma_deriv / float(N)) * learning_rate
    alpha -= (alpha_deriv / float(N)) * learning_rate
    beta -= (beta_deriv / float(N)) * learning_rate
    return gamma, alpha, beta



def fit_polinomial_quadr(T, X, gamma, alpha, beta, learning_rate, iters):
    cost_history = []
    space_inter=iters/5
    mse_fim=0
    rmse_fim=0
    for i in range(int(iters)):
        gamma,alpha,beta = update_weights_pol_quadr(T, X, gamma, alpha, beta, learning_rate)

        #Calculate cost for auditing purposes
        mse_ini,rmse_ini,x_hat = cost_function_pol_quadr(T, X, gamma, alpha, beta)
        cost_history.append(rmse_ini)
        
        #Checking for the tolerance of the error
        tol_rmse = np.abs(rmse_ini-rmse_fim)
        tol_mse = np.abs(mse_ini-mse_fim)
        if tol_rmse < 1e-10:
            break
        mse_fim=mse_ini
        rmse_fim=rmse_ini
        
        # Log Progress
        if i % space_inter == 0:
            print ("iter={:d}  gamma={:.2f}  alpha={:.2f}    beta={:.4f}    MSE={:.2}     RMSE={:.2}".format(i,gamma, alpha, beta, mse_fim, rmse_fim))
            plt.scatter(T,X)
            plt.plot(T,x_hat,c='red')
            plt.show()
    return rmse_fim, gamma, alpha, beta, cost_history, i
```
